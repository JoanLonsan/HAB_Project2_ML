```python
import pandas as pd
from joblib import dump

from sklearn.ensemble import RandomForestClassifier, ExtraTreesRegressor
```

```python
df = pd.read_csv('../4.3 csv_prep_ml/df_ml.csv')
target_clas = 'next_class_encoded'
target_reg = 'next_score'
```

```python
df
```

```python
X = df.drop(columns = [target_clas, target_reg], axis = 1)
y = df[target_clas]
```

```python
model = RandomForestClassifier(criterion = 'entropy',
                               max_depth = 35,
                               max_features = None,
                               min_samples_leaf = 10,
                               n_estimators = 200)

model.fit(X, y)
```

```python
dump(model, './model_classification.joblib')
```

```python
X = df.drop(columns = target_reg, axis = 1)
y = df[target_reg]
```

```python
model = ExtraTreesRegressor(max_depth = None,
                            max_features = None,
                            min_samples_leaf = 1,
                            min_samples_split = 2,
                            n_estimators = 200)

model.fit(X, y)
```

```python
dump(model, './model_regression.joblib')
```
