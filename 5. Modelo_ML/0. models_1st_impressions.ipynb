```python
# Sklearn tiene las formulas de algunas métricas en funciones.

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import root_mean_squared_error
```

```python
# Relative Absolute Error

RAE = np.sum(np.abs(np.subtract(y_tst, yhat_reg))) / np.sum(np.abs(np.subtract(y_tst, np.mean(y_tst))))
```

```python
# Relative Square Error

RSE = np.sum(np.square(np.subtract(y_tst, yhat_reg))) / np.sum(np.square(np.subtract(y_tst, np.mean(y_tst))))
```

```python
r2_ajustada = 1 - (1 - model.score(X_tst, y_tst))*(len(y_tst) - 1)/(len(y_tst) - X_tst.shape[1] - 1)
```

```python
print(f"MAE:\t {mean_absolute_error(y_tst, yhat_reg)}")
print(f"MSE:\t {mean_squared_error(y_tst, yhat_reg)}")
print(f"RMSE: {root_mean_squared_error(y_tst, yhat_reg)}")
print(f"R**2:\t {r2_score(y_tst, yhat_reg)}")
print(f"RAE:\t {RAE}")
print(f"RSE:\t {RSE}")
print(f"Adjusted R**2:\t {r2_ajustada}")
```
