```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

from joblib import dump
```

```python
df_raw = pd.read_csv("../2.3 csv_unique/df_unified.csv")
df_raw.head(20)
```

```python
df_raw = df_raw.sort_values(by=['names', 'dates'])

# Crear la columna "next_class" tomando el valor de "calidad_aire" del siguiente día para cada estación
df_raw['next_class'] = df_raw.groupby('names')['calidad_aire'].shift(-1)
# Lo mismo con next_score
df_raw['next_score'] = df_raw.groupby('names')['scores'].shift(-1)
```

```python
df_raw.info()
```

```python
df_raw['names'].unique()
```

```python
df_raw['dates'] = pd.to_datetime(df_raw['dates'], errors='coerce')
df_raw['Dia'] = pd.to_datetime(df_raw['Dia'])
df_raw['scores'] = pd.to_numeric(df_raw['scores'], errors='coerce')
df_raw['next_score'] = pd.to_numeric(df_raw['next_score'], errors='coerce')
```

```python
# Definir cols categoricas y numericas
categorical_features = ['names', 'comunity', 'periodo', 'calidad_aire']
numeric_features = df_raw.select_dtypes(include=['float64', 'int64']).columns.tolist()
```

```python
numeric_features.remove('scores')
numeric_features.remove('next_score')
```

```python
numeric_features
```

```python
df_raw.describe()
```

```python
# Reducir MemoryUsage
df_raw['scores'] = df_raw['scores'].astype(np.float16)
df_raw['co'] = df_raw['co'].astype(np.float32)
df_raw['humidity'] = df_raw['humidity'].astype(np.float32)
df_raw['no2'] = df_raw['no2'].astype(np.float32)
df_raw['o3'] = df_raw['o3'].astype(np.float32)
df_raw['p'] = df_raw['p'].astype(np.float32)
df_raw['pm10'] = df_raw['pm10'].astype(np.float32)
df_raw['pm25'] = df_raw['pm25'].astype(np.float32)
df_raw['so2'] = df_raw['so2'].astype(np.float32)
df_raw['temp'] = df_raw['temp'].astype(np.float32)
df_raw['wind'] = df_raw['wind'].astype(np.float32)
df_raw['Hora'] = df_raw['Hora'].astype(np.int8)
df_raw['next_score'] = df_raw['next_score'].astype(np.float16)
```

```python
for col in categorical_features:
    df_raw[col] = df_raw[col].astype('category')
```

```python
df_raw['calidad_aire'].unique()
```

```python
# Ajustar y transformar las columnas con map para personalizar valores, no lo hago con Label encoder para personalizar los numeros con sus valores, y dejar el valor mas alto como el equivalente el Nan o que hay que repasar
periodo_mapping = {'mañana': 0, 'mediodia': 1, 'tarde': 2, 'noche': 3, 'otro': 4}
calidad_aire_mapping = {'Muy Dañina a la Salud': 0, 'Dañina a la Salud': 1, 'Dañina a la Salud de los Grupos Sensitivos': 2, 'Moderada': 3, 'Buena': 4, '-': 5}
df_raw['periodo_encoded'] = df_raw['periodo'].map(periodo_mapping)
df_raw['calidad_aire_encoded'] = df_raw['calidad_aire'].map(calidad_aire_mapping)
df_raw['next_class_encoded'] = df_raw['next_class'].map(calidad_aire_mapping)
```

```python
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_cats = encoder.fit(df_raw[categorical_features])
```

```python
imputer = IterativeImputer()
imputed_nums = imputer.fit(df_raw[numeric_features])
```

```python
dump(imputer, '../4.2 models_prepro/imputer.joblib')
print("Imputer guardados en '4.2 models_prepro'")
```

```python
imputer_nums_df = pd.DataFrame(imputed_nums.transform(df_raw[numeric_features]), columns=numeric_features)
```

```python
scaler = StandardScaler()
scaled_nums = scaler.fit(imputer_nums_df)
```

```python
dump(encoder, '../4.2 models_prepro/encoder.joblib')
dump(scaler, '../4.2 models_prepro/scaler.joblib')
print("Encoder, Scaler e Imputer guardados en '4.2 models_prepro'")
```

```python
encoded_cats_df = pd.DataFrame(encoded_cats.transform(df_raw[categorical_features]), columns=encoder.get_feature_names_out(categorical_features))
```

```python
scaled_nums_df = pd.DataFrame(scaled_nums.transform(imputer_nums_df[numeric_features]), columns=numeric_features)
```

```python
add_cols_df = pd.concat([df_raw['scores'], df_raw['next_score'], df_raw['periodo_encoded'], df_raw['calidad_aire_encoded'], df_raw['next_class_encoded']], axis=1)
```

```python
add_cols_df
```

```python
df_preprocessed = pd.concat([add_cols_df, scaled_nums_df, encoded_cats_df], axis=1)
```

```python
df_preprocessed = df_preprocessed.dropna(subset=['next_score'])
df_preprocessed = df_preprocessed.dropna(subset=['scores'])
df_preprocessed
```
