```python
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv('../1.2 csvs_raw/cities_stations_20240625_223537.csv')
```

```python
df['dates'] = pd.to_datetime(df['dates'])
df['fecha_convertida'] = df['dates'] - pd.Timedelta(hours=7)

# Crear las nuevas columnas 'Dia' y 'Hora'
df['Dia'] = df['fecha_convertida'].dt.date
df['Hora'] = df['fecha_convertida'].dt.hour

print(df)
```

## SEPARAR FECHA Y HORA Y CATEGORIZAR ESTA ULTIMA

```python
# Función para categorizar la hora
def categorizar_hora(hora):
    if 6 <= hora < 12:
        return 'mañana'
    elif 12 <= hora < 15:
        return 'mediodia'
    elif 15 <= hora < 20:
        return 'tarde'
    elif 20 <= hora < 24:
        return 'noche'
    else:
        return 'otro'

# Aplicar la función a la columna 'hora'
df['periodo'] = df['Hora'].apply(categorizar_hora)

# Eliminar la columna 'hora' original, ya que no se necesita más
# df = df.drop(columns=['hora'])
```

## CATEGORIZAR CALIDAD DEL AIRE

```python
def categorizar_calidad_aire(score):
    if score <= 50:
        return 'Buena'
    elif score <= 100:
        return 'Moderada'
    elif score <= 150:
        return 'Dañina a la Salud de los Grupos Sensitivos'
    elif score <= 200:
        return 'Dañina a la Salud'
    elif score <= 300:
        return 'Muy Dañina a la Salud'
    else:
        return 'Arriesgado'
```

```python
# Aplicar la función a la columna 'scores'
df['calidad_aire'] = df['scores'].apply(lambda x: categorizar_calidad_aire(float(x)) if x != '-' else '-')
```

```python
# Reordenar las columnas para que 'fecha' y 'periodo' estén juntas
df = df[['names', 'comunity', 'scores','calidad_aire', 'fecha_convertida', 'periodo', 'co', 'humidity', 'no2', 'o3', 'p', 'pm10', 'pm25', 'so2', 'temp', 'wind']]
```
