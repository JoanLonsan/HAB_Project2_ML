```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../1.2 csvs_raw/cities_stations_20240625_223537.csv")
df.head(10)
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['scores'], kde=True)
plt.title('Distribución de Scores')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['pm10'], kde=True)
plt.title('Distribución de partículas inferior a 10 micrometros')
plt.show()
```

## Heatmap de correlaciones

```python
df['scores'] = pd.to_numeric(df['scores'], errors='coerce')


# Seleccionamos solo las columnas numéricas
df_numerico = df.select_dtypes(include=[np.number])


# Calculamos la correlación
corr = df_numerico.corr()


# Creamos el mapa
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, fmt=".1f", cmap='viridis')
plt.title("Correlación de valores")
plt.show()
```

## Graficas de pares

```python
sns.pairplot(data= df)
plt.title("gráfica de pares")
plt.show()
```

```python
# Análisis bivariado: Scores vs Humidity
plt.figure(figsize=(8, 4))
sns.scatterplot(data=df, x='pm10', y='pm25')
plt.title('Partículas en suspensión de 10 micrometros vs 25 micrometros')
plt.show()
```

```python
# Análisis de tendencias temporales
plt.figure(figsize=(8, 4))
df.set_index('dates')['temp'].plot()
plt.title('Tendencia de Temperatura a lo largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Temperatura')
plt.show()
```

```python
# Análisis bivariado: Scores vs Humidity
plt.figure(figsize=(8, 4))
sns.scatterplot(data=df, x='humidity', y='scores')
plt.title('Scores vs Humidity')
plt.show()
```

```python
# Análisis bivariado: Scores vs Ozone
plt.figure(figsize=(8, 4))
sns.scatterplot(data=df, x='o3', y='scores')
plt.title('Scores vs Ozone')
plt.show()
```

```python
# Gráfico de enjambre
sns.swarmplot(data=df, x="co", y="pm10", hue="dates")
plt.legend([], [], frameon=False)
plt.title("Swarm Plot: Total Bill by Day and Time")
plt.show()
```

```python
sns.histplot(data=df, x="pm10", hue="dates", multiple="stack")
plt.legend([], [], frameon=False)
plt.title("Partículas a lo largo del tiempo")
plt.show()
```
