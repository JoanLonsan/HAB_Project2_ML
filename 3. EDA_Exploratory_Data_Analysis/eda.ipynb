```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../1.2 csvs_raw/cities_stations_20240625_223537.csv")
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['scores'], kde=True)
plt.title('Distribucin de Scores')
plt.show()
```

```python
# Correlacin entre variables
corr_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlacin')
plt.show()
```

```python
# Anlisis de tendencias temporales
plt.figure(figsize=(10, 6))
df.set_index('dates')['temp'].plot()
plt.title('Tendencia de Temperatura a lo largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Temperatura')
plt.show()
```

```python
# Anlisis bivariado: Scores vs Humidity
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='humidity', y='scores')
plt.title('Scores vs Humidity')
plt.show()
```
